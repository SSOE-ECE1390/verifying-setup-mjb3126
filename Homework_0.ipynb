* Edit brightness of image
* Rotate image

```python
import cv2, numpy as np, matplotlib.pyplot as plt

apollo = cv2.imread("data/Apollo_11_Launch.jpg")
#img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
cropped_apollo = apollo[175:400, 400:800]

# Show the images
plt.figure(figsize=[18, 5])
plt.subplot(131); plt.imshow(apollo);  plt.title("Loaded Image");
plt.subplot(132); plt.imshow(cropped_apollo);  plt.title("Cropped");
```

```python
matrix = np.ones(apollo.shape, dtype="uint8") * 100
apollo_brighter = cv2.add(apollo, matrix)
apollo_rotated = cv2.rotate(apollo, cv2.ROTATE_90_CLOCKWISE)
# Show the images
plt.figure(figsize=[18, 5])
plt.subplot(131); plt.imshow(apollo_brighter);  plt.title("Brighter");
plt.subplot(132); plt.imshow(apollo_rotated);  plt.title("Rotated");

cv2.imwrite("data/apollo_rotated.png", apollo_rotated)
```
